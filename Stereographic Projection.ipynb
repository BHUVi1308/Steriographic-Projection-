#### Code(formulae) to convert cartesian co-ordinate into projection on plane as got from wikipedia page

 Here plane refers to plane coordinate and (x1,y1.z1) refers to cartesian co-ordinate
 
 The function change converts cartesian co-ordinates to projection on plane

```python
def change(x1,y1,z1,plane):
    plane[0] = x1/(1-z1)
    plane[1] = y1/(1-z1)
    return
```

#### To show how this works I will take a point on sphere in cartesian co-ordinate and convert it into steriographic projection

```python
plane = [0,0]
x1= 0
y1= (3/4)^(1/2)
z1= 1/2
change(x1,y1,z1,plane)
```

#### Code to convert a point specified using theta and phi in spherical co-ordinate to cartesian co-ordinate

Here 0<theta<2pi and 0<phi<pi
 
So points in spherical co-ordinate is specified it can be converted into points on cartesian co-ordinate and can be proceeded

Convert is name of the function used to achieve this 

```python
import random
def convert(sph,theta,phi,r):
    sph[0] = r*cos(theta) * sin(phi)
    sph[1] = r*sin(theta) * sin(phi)
    sph[2] = r*cos(phi)
    return
```

#### Here I am taking an example of converting spherical co-ordinates to projection on plane

```python
sph= [0,0,0]
plane=[0,0]
convert(sph,math.pi/6,2*math.pi/3,1)
change(sph[0],sph[1],sph[2],plane)
```

#### Now to the next part of the question projection of line on the sphere

Here I've considered line as an array of points which is assumption that can be made according to the question

```python
import numpy as np
theta = np.linspace(0,2*math.pi,20)
phi = np.linspace(math.pi/3,2*math.pi/3,20)
```

#### As an example of line on the sphere I am taking line by considering phi as variable and theta as constant which gives a curve on the sphere

This program here converts each point of the array to a point in the steriographic projection and plots in the graph

```python
sp = []
pp = []

for i in range(0,20):
    sph = [0,0,0]
    plane = [0,0]
    convert(sph,theta[0],phi[i],1)       
    change(sph[0],sph[1],sph[2],plane)
    sp.append(point((sph[0],sph[1],sph[2]), marker = 'o', color ='black', size = 10))
    pp.append(point((plane[0],plane[1],0), marker = 'o', color ='black', size = 10))
```

```python
pns = implicit_plot3d(x^2+y^2+z^2==1, (x,-2,2), (y,-2,2), (z,-2,2), color = "red" , opacity = 0.3, smooth = True,viewer='tachyon')
pns += implicit_plot3d(z==0, (x,-2,2), (y,-2,2), (z,-2,2), color = 'white', opacity = 0.5,viewer='tachyon')
for i in range(0,20):
    pns += sp[i]
    pns += pp[i]
    
pns.show()
```
